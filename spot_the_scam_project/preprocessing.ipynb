```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import os

# Create output directory if it doesn't exist
os.makedirs("data", exist_ok=True)

# Load the training and test data
train_df = pd.read_csv("data/train.csv")
test_df = pd.read_csv("data/test.csv")

# Function to clean text
def clean_text(text):
    if pd.isna(text):
        return ""
    text = str(text).lower()
    return text.strip()

# Apply text cleaning to relevant columns
for col in ['title', 'description', 'company_profile', 'requirements', 'benefits']:
    if col in train_df.columns:
        train_df[col] = train_df[col].apply(clean_text)
    if col in test_df.columns:
        test_df[col] = test_df[col].apply(clean_text)

# Fill missing values with 'Unknown' or appropriate defaults
train_df.fillna("Unknown", inplace=True)
test_df.fillna("Unknown", inplace=True)

# Label encode categorical columns common to both train and test
cat_cols = ['employment_type', 'required_experience', 'required_education', 'industry', 'function']

for col in cat_cols:
    if col in train_df.columns and col in test_df.columns:
        le = LabelEncoder()
        le.fit(list(train_df[col]) + list(test_df[col]))
        train_df[col] = le.transform(train_df[col])
        test_df[col] = le.transform(test_df[col])

# Save processed CSVs
train_df.to_csv("data/processed_train.csv", index=False)
test_df.to_csv("data/processed_test.csv", index=False)
```
